```python
import pandas as pd
import numpy as np
```

```python
merch=pd.read_csv('DATA Leroy Merlin.xlsx - Données brutes MERCH 22 23.csv')
merch
```

```python
merch_2022=merch.query('`ANNEE `== 2022') 
```

```python

merch_2023=merch.query('`ANNEE `== 2023') 
merch_2023
merch_2023.columns
merch_2023.shape
print(merch_2023['BUDGET ESTIMÉ'].apply(type).value_counts())

```

```python
merch_2023['BUDGET ESTIMÉ']=merch_2023['BUDGET ESTIMÉ'].str.replace(' €','').str.replace(' ','').str.replace(',','.').str.replace('\u202f','')
merch_2023['BUDGET ESTIMÉ'].describe
merch_2023['BUDGET ESTIMÉ'].astype(float)
```

```python
sales=pd.read_csv('DATA Leroy Merlin.xlsx - CA SELL IN LM 22 23.csv')
sales
```

```python
sales_tot=sales[sales['Total\nFamille']=='TOTAL HENKEL']
sales_tot
sales_tot.columns
sales_tot.shape
```

```python
merged_2023=pd.merge(sales_tot,merch_2023,left_on='Raison Sociale',right_on='RAISON SOCIALE',how='right')
merged_2023
```

```python
merged_2023=merged_2023[['Code du Point de Vente', 'Raison Sociale','Total\nannuel\n2023', 'PASSAGE','HEURES','BUDGET ESTIMÉ']]
```

```python
merged_2023=merged_2023.rename(columns={"Total\nannuel\n2023" : "Total sales annuel", "BUDGET ESTIMÉ" : "BUDGET ESTIMÉ MERCH"})

```

```python
merged_2023["HEURES"].describe()
```

```python
# Conversion en heures décimales
merged_2023['HEURES'] = merged_2023['HEURES'].str.extract(r'(\d+)h(\d+)').astype(float).apply(lambda x: x[0] + x[1] / 60, axis=1)

merged_2023


```

```python


merged_2023['HEURES/SEMAINE'] = np.where(merged_2023['PASSAGE'] == 'Quinzaine', 2 * merged_2023['HEURES'], merged_2023['HEURES'])
merged_2023
merged_2023["BUDGET ESTIMÉ MERCH"].iloc[79:100]
```

```python
visits=pd.read_csv('DATA Leroy Merlin.xlsx - VISITES ANNUELLES 22 23.csv')
visits
```

```python
visits_2023=visits[['CODE INTERNE UNIQUE DU POINT DE VENTE','Total visites annuelles\n2023']]
visits_2023=visits_2023.rename(columns={"Total visites annuelles\n2023": "Total visites annuelles"})
visits_2023
```

```python

merged_2023=pd.merge(merged_2023,visits_2023,left_on='Code du Point de Vente',right_on='CODE INTERNE UNIQUE DU POINT DE VENTE',how='inner')

merged_2023['BUDGET ESTIMÉ MERCH'].describe()
```

```python
merged_2023.drop(columns =["CODE INTERNE UNIQUE DU POINT DE VENTE"], inplace = True)

merged_2023
```

```python

merged_2023_final= merged_2023.drop(columns=['HEURES'])
merged_2023_final
```
